```python
# Dependencies and Setup, import the data set
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
#Just to see the data we are working with here
purchase_data
```

## Player Count

* Display the total number of players


```python
#define total players and display
#check the data and look for duplicates
total_players = purchase_data["SN"]
total_players = total_players.drop_duplicates()
total_players = total_players.count()
#set up table that will show this total, set up a function to show and use pd.DataFrame
pd.DataFrame({'Total Players': [total_players]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Use the data to summarize the data from the total of 780
#set up variables and solve each piece in the requested data frame
unique_items = purchase_data["Item ID"].nunique() #nunique function removes NA variables and counts

#find average price

average_price = (purchase_data["Price"].mean())

#round average price to only 2 places
rounded_average_price = round(average_price, 2)

#find total number of purchases
number_of_purchases = purchase_data["Purchase ID"].count()

#find total revenue
total_revenue = purchase_data["Price"].sum()
```

```python
#Show variables in data frame:
summaries_df = pd.DataFrame([{"Number of Unique Items": unique_items, "Average Price": rounded_average_price, "Number of Purchase": number_of_purchases, "Total Revenue": total_revenue}])
#add the dollar sign using .map format
summaries_df["Average Price"] = summaries_df["Average Price"].map("${:.2f}".format)
summaries_df["Total Revenue"] = summaries_df["Total Revenue"].map("${:.2f}".format)
#set up order of columns
summaries_df = summaries_df[["Number of Unique Items", "Average Price", "Number of Purchase", "Total Revenue"]]
summaries_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
groupedby = purchase_data.groupby(['Gender'])
gender_counts = groupedby['SN'].nunique()
sum_gender_counts = gender_counts.sum()

male_count = gender_counts['Male']
male_percent = (male_count/sum_gender_counts) * 100

female_count = gender_counts['Female']
female_percent = (female_count/sum_gender_counts) * 100


other_count = gender_counts['Other / Non-Disclosed']
other_percent = (other_count/sum_gender_counts) * 100

gender_df = pd.DataFrame(gender_counts)
gender_df['Percents'] = [male_percent, female_percent, other_percent]
gender_df["Percents"] = gender_df["Percents"].map("{:.2f}".format)
gender_df.columns = ["Total Count", "Percentage"]

gender_df

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
total_item_count_by_gender = purchase_data.groupby('Gender')['Item ID'].count()
total_spent_by_gender = purchase_data.groupby('Gender')['Price'].sum()
max_cost_by_gender = purchase_data.groupby('Gender')['Price'].max()
min_cost_by_gender = purchase_data.groupby('Gender')['Price'].min()
avg_cost_by_gender = purchase_data.groupby('Gender')['Price'].mean()

avg_spent_by_female = total_spent_by_gender.Female / gender_counts.Female
avg_spent_by_male = total_spent_by_gender.Male / gender_counts.Male
avg_spent_by_other = total_spent_by_gender["Other / Non-Disclosed"] / gender_counts["Other / Non-Disclosed"]

myList = [{"Total Items": total_item_count_by_gender.Female, "Total Purchases": total_spent_by_gender.Female,
            "Avg Spent per Player": avg_spent_by_female, "Average Purchase Price": avg_cost_by_gender.Female},
           {"Total Items": total_item_count_by_gender.Male, "Total Purchases": total_spent_by_gender.Male,
            "Avg Spent per Player": avg_spent_by_male, "Average Purchase Price": avg_cost_by_gender.Male},
           {"Total Items": total_item_count_by_gender["Other / Non-Disclosed"],
            "Total Purchases": total_spent_by_gender["Other / Non-Disclosed"], "Avg Spent per Player": avg_spent_by_other,
            "Average Purchase Price": avg_cost_by_gender["Other / Non-Disclosed"]}]

gender_totals_df = pd.DataFrame(myList)
gender_totals_df.rename(index={0:'Female', 1:'Male', 2:'Other'}, inplace=True)

gender_totals_df["Average Purchase Price"] = gender_totals_df["Average Purchase Price"].map("${:.2f}".format)
gender_totals_df["Total Purchases"] = gender_totals_df["Total Purchases"].map("${:.2f}".format)
gender_totals_df["Avg Spent per Player"] = gender_totals_df["Avg Spent per Player"].map("${:.2f}".format)
# Change column order
gender_totals_df = gender_totals_df[['Average Purchase Price','Total Items', 'Total Purchases', 'Avg Spent per Player']]

gender_totals_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)

groupedby_ages = purchase_data.groupby('Age Ranges')
age_counts = groupedby_ages['SN'].nunique()   
total_unique_players = age_counts.sum()

age_list = [{"Total Count": age_counts["<10"], "Percentage": (age_counts["<10"] * 100) / total_unique_players},
            {"Total Count": age_counts["10-14"], "Percentage": (age_counts["10-14"] * 100) / total_unique_players},
            {"Total Count": age_counts["15-19"], "Percentage": (age_counts["15-19"] * 100) / total_unique_players},
            {"Total Count": age_counts["20-24"], "Percentage": (age_counts["20-24"] * 100) / total_unique_players},
            {"Total Count": age_counts["25-29"], "Percentage": (age_counts["25-29"] * 100) / total_unique_players},
            {"Total Count": age_counts["30-34"], "Percentage": (age_counts["30-34"] * 100) / total_unique_players},
            {"Total Count": age_counts["35-39"], "Percentage": (age_counts["35-39"] * 100) / total_unique_players},
            {"Total Count": age_counts["40+"], "Percentage": (age_counts["40+"] * 100) / total_unique_players}]

summary_ages_df = pd.DataFrame(age_list)
summary_ages_df.rename(index={0:'<10', 1:'10-14', 2:'15-19', 3:'20-24', 4:'25-29', 5:'30-34', 6:'35-39', 7:'40+'}, inplace=True)
summary_ages_df["Percentage"] = summary_ages_df["Percentage"].map("{:.2f}%".format)
summary_ages_df = summary_ages_df[["Total Count", "Percentage"]]

summary_ages_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
groupedby_player = purchase_data.groupby(['SN'])
total_spent_player = groupedby_player['Price'].sum()
items_per_player = groupedby_player['Price'].count()

player_list_df = pd.DataFrame(total_spent_player)
player_list_df['Purchase Count'] = items_per_player
player_list_df['Average Purchase Price'] = total_spent_player / items_per_player
player_list_df.columns = ["Total Purchase Value", "Purchase Count", "Average Purchase Price"]
player_list_df
# Formatting
player_list_df["Average Purchase Price"] = player_list_df["Average Purchase Price"].map("${:.2f}".format)
player_list_df.sort_values(by='Total Purchase Value', ascending=False, inplace=True)
player_list_df["Total Purchase Value"] = player_list_df["Total Purchase Value"].map("${:.2f}".format)

player_list_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_item_df = purchase_data.loc[:, 'Item ID':'Price'].copy()
groupedby_popular = popular_item_df.groupby(['Item ID', 'Item Name'])

total_purchases = groupedby_popular['Price'].sum()
num_items_purchased = groupedby_popular['Item ID'].count()

# Create dataframe from groupby values
popular_df = pd.DataFrame(num_items_purchased)
popular_df['Total Purchase Value'] = total_purchases
popular_df.columns = ["Purchase Count", "Total Purchase Value"]
popular_df = popular_df.sort_values(by='Item ID')

# Create prices dataframe
prices_df = purchase_data[['Item ID', 'Item Name', 'Price']]
prices_nodup_df = prices_df.drop_duplicates(['Item ID'], keep = 'last')
prices_post_df = prices_nodup_df.sort_values(by=['Item ID'])
prices_post_df.set_index(['Item ID', 'Item Name'], inplace=True)
# Join the 2 dataframes
popular_df = popular_df.join(prices_post_df)

# Copy for most profitable list below because formatting screws up sorting!)
profit_df = popular_df

# Sort by Purchase Count and format
popular_df = popular_df.sort_values(by='Purchase Count', ascending=False)
popular_df["Price"] = popular_df["Price"].map("${:.2f}".format)
popular_df["Total Purchase Value"] = popular_df["Total Purchase Value"].map("${:.2f}".format)

popular_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Sort by Total Purchase Value and format
profit_df = profit_df.sort_values(by='Total Purchase Value', ascending=False)
profit_df["Price"] = profit_df["Price"].map("${:.2f}".format)
profit_df["Total Purchase Value"] = profit_df["Total Purchase Value"].map("${:.2f}".format)

profit_df.head()
```
